```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def compressImage(img):
    r = img[: , : , 0]
    g = img[: , : , 1]
    b = img[: , : , 2]
    
    u_r, s_r, vt_r = np.linalg.svd(r)
    s_r = np.diag(s_r)

    u_g, s_g, vt_g = np.linalg.svd(g)
    s_g = np.diag(s_g)

    u_b, s_b, vt_b = np.linalg.svd(b)
    s_b = np.diag(s_b)
    
    i = 0
    for k in (10, 50, 150, 250, 500):
        
        a_r = u_r[:,0:k] @ s_r[0:k,0:k] @ vt_r[0:k,:]

        a_g = u_g[:,0:k] @ s_g[0:k,0:k] @ vt_g[0:k,:]

        a_b = u_b[:,0:k] @ s_b[0:k,0:k] @ vt_b[0:k,:]

        a = np.zeros((len(img),len(img[0]),3))

        a[:, : , 0] = a_r
        a[:, : , 1] = a_g
        a[:, : , 2] = a_b

        i += 1
        plt.figure(i)
        plt.imshow(a.astype('uint8'))
        plt.axis('off')
        plt.title('k = ' + str(k))
        plt.show()
```

```python
# compressing the first image
img1 = plt.imread("images/1.bmp")
plt.imshow(img1)
plt.show()
```

```python
compressImage(img1)
```

```python
# compressing the second image
img2 = plt.imread("images/2.bmp")
plt.imshow(img2)
plt.show()
```

```python
compressImage(img2)
```

```python
# compressing the third image
img3 = plt.imread("images/3.bmp")
plt.imshow(img3)
plt.show()
```

```python
compressImage(img3)
```
